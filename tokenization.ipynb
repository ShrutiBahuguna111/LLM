**##Step 1: Tokenizing the short story**

```python
# Open the file named "the-verdict.txt" in read mode ("r")
# encoding="utf-8" helps Python read text containing special characters
with open("the-verdict.txt", "r", encoding="utf-8") as f:

    # Read the entire content of the file and store it in the variable raw_text
    raw_text = f.read()

# len(raw_text) returns the total number of characters in the file
print("Total number of characters:", len(raw_text))

# Display the first 99 characters of the text
# [:99] is called slicing and extracts characters from index 0 to 98
print(raw_text[:99])
```

```python
# Import the regular expression (re) module
import re

# A sample text string
text = "Hello, world. This, is a test."

# Split the text wherever a whitespace character (\s) is found
# Parentheses () keep the whitespace characters in the output list
result = re.split(r'(\s)', text)

# Print the resulting list
print(result)
```

```python
# Create a new list by keeping only non-empty items
# item.strip() removes spaces from the beginning and end of the string
# If the result is not empty, the item is kept in the new list
result = [item for item in result if item.strip()]
# Print the filtered list
print(result)
```

```python
# Sample text containing words, punctuation marks, and spaces
text = "Hello, world. Is this-- a test?"

# Split the text based on:
# - punctuation marks: , . : ; ? _ ! " ( ) '
# - double hyphen: --
# - whitespace characters: \s
# Parentheses () keep the matched separators in the output list
result = re.split(r'([,.:;?_!"()\']|--|\s)', text)

# Remove extra spaces from each item using strip()
# Keep only non-empty items
result = [item.strip() for item in result if item.strip()]

# Print the final list
print(result)
```

```python
# Create a new list containing only non-empty items
# item.strip() removes any spaces before and after the text
# If the item becomes an empty string after strip(), it is not included
result = [item for item in result if item.strip()]

# Display the filtered list
print(result)
```

```python
# Split the text into smaller parts (tokens)
# The text will be split whenever Python finds:
# - punctuation marks (, . : ; ? _ ! " ( ) ')
# - a double hyphen (--)
# - a whitespace character (space, tab, newline)
preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text)

# Remove extra spaces from each token using strip()
# Keep only non-empty tokens
preprocessed = [item.strip() for item in preprocessed if item.strip()]

# Display the first 30 tokens from the processed text
print(preprocessed[:30])
```

**## Step 2: Creating Token IDs**

```python
# Convert the list of tokens into a set to remove duplicate words/tokens
# Then sort them alphabetically and store them in all_words
all_words = sorted(set(preprocessed))

# Count the total number of unique words/tokens in the vocabulary
vocab_size = len(all_words)

# Display the vocabulary size
print(vocab_size)
```

```python
# Create a dictionary called vocab
# enumerate(all_words) assigns a number (index) to each token
# Format: {token : index}
vocab = {token: integer for integer, token in enumerate(all_words)}
```

```python
# Loop through the dictionary items along with their position
for i, item in enumerate(vocab.items()):

    # Print each (token, index) pair
    print(item)

    # Stop after printing the first 51 items (0 to 50)
    if i >= 50:
        break
```

```python
# Define a simple tokenizer class
class SimpleTokenizerV1:

    # Constructor: runs automatically when an object is created
    def __init__(self, vocab):

        # Dictionary that maps tokens (words/punctuation) to integers
        self.str_to_int = vocab

        # Create a reverse dictionary that maps integers back to tokens
        self.int_to_str = {i: s for s, i in vocab.items()}

    # Convert text into token IDs
    def encode(self, text):

        # Split text into tokens based on punctuation, double hyphens, and spaces
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)

        # Remove extra spaces and empty strings
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]

        # Convert each token into its corresponding integer ID
        ids = [self.str_to_int[s] for s in preprocessed]

        # Return the list of token IDs
        return ids

    # Convert token IDs back into text
    def decode(self, ids):

        # Convert each ID back to its token and join them with spaces
        text = " ".join([self.int_to_str[i] for i in ids])

        # Remove unwanted spaces before punctuation marks
        # Example: "Hello , world !" -> "Hello, world!"
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)

        # Return the reconstructed text
        return text
```

```python
# Create an object of the SimpleTokenizerV1 class
# The vocabulary dictionary (vocab) is passed to the tokenizer
tokenizer = SimpleTokenizerV1(vocab)

# Sample text to be converted into token IDs
text = """"It's the last he painted, you know,"
           Mrs. Gisburn said with pardonable pride."""

# Convert the text into a list of integer IDs
ids = tokenizer.encode(text)

# Display the encoded token IDs
print(ids)
```

```python
# Display the decoded token
tokenizer.decode(ids)
```

**## Step 3: ADDING SPECIAL CONTEXT TOKENS**


```python
# Create a set from preprocessed tokens to remove duplicates
# Then convert it back to a list and sort it alphabetically
all_tokens = sorted(list(set(preprocessed)))

# Add two special tokens to the vocabulary
# <|endoftext|> : marks the end of a text/document
# <|unk|>       : represents an unknown word not found in the vocabulary
all_tokens.extend(["<|endoftext|>", "<|unk|>"])

# Create a dictionary mapping each token to a unique integer ID
vocab = {token: integer for integer, token in enumerate(all_tokens)}
```

```python
# Define an improved tokenizer class
class SimpleTokenizerV2:

    # Constructor: initializes the tokenizer with a vocabulary
    def __init__(self, vocab):

        # Dictionary to convert tokens (words) into IDs
        self.str_to_int = vocab

        # Reverse dictionary to convert IDs back into tokens
        self.int_to_str = {i: s for s, i in vocab.items()}

    # Convert text into token IDs
    def encode(self, text):

        # Split the text into words and punctuation marks
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)

        # Remove spaces and empty strings
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]

        # If a token is not present in the vocabulary,
        # replace it with the special token <|unk|>
        preprocessed = [
            item if item in self.str_to_int
            else "<|unk|>"
            for item in preprocessed
        ]

        # Convert each token into its corresponding ID
        ids = [self.str_to_int[s] for s in preprocessed]

        # Return the list of IDs
        return ids

    # Convert IDs back into readable text
    def decode(self, ids):

        # Convert each ID back to its token
        # and join them using spaces
        text = " ".join([self.int_to_str[i] for i in ids])

        # Remove unnecessary spaces before punctuation
        # Example: "Hello , world ." → "Hello, world."
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)

        # Return the reconstructed text
        return text
```

```python
tokenizer = SimpleTokenizerV2(vocab)

text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."

text = " <|endoftext|> ".join((text1, text2))

print(text)
```

```python
tokenizer.encode(text)
```

```python
tokenizer.decode(tokenizer.encode(text))
```
